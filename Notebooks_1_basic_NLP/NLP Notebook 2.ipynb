```python
import gensim
from gensim.models import Word2Vec, KeyedVectors
import gensim.downloader as api
```

```python
# Google News word2vec pre-trained model
# ======================================
# Window size = 300
# Architechture = Skipgram
# Output Dimension = 300
#
# Loading the Google News word2vec pre-trained model
# --------------------------------------------------
wv=api.load('word2vec-google-news-300') 
```

```python
# Printing the model
print(wv)
```

```python
# Checking the vector for "king"
vector_king = wv['king']
vector_king     # This will be a vector array of size = 300
```

```python
# Checking the vector for "man"
vector_man = wv['man']
print(vector_man[0:50]) # Printing first 50 of 300 values
```

```python
# Checking most similar words based on a given word
# -------------------------------------------------
# Criteria = Cosine Similarity
#
# Checking for "man"
wv.most_similar('man')
```

```python
# Checking most similar words for "king"
wv.most_similar('king')
```

```python
# Finding similarity bewtween 2 words
# -----------------------------------
# Criteria = Cosine Similarity
#
print(f"Similarity between man and king is          :{wv.similarity('man', 'king')}")
print(f"Similarity between HTML and programmer is   :{wv.similarity('html', 'programmer')}")
print(f"Similarity between hockey and sport is      :{wv.similarity('hockey', 'sport')}")
```

```python
# We can also do some operations and find a resulting vector
# ----------------------------------------------------------
vec1 = wv['king']-wv['man']+wv['woman']
wv.most_similar([vec1])
```
